```python
def filmografia(actor:str):#define funcion filmografia que recibe una cadena de texto
    result1 = net_peliculas[net_peliculas['cast'].str.contains(actor,na=False)]#filtra registros que contenga la cadena de texto ingresada
    result2 = net_series[net_series['cast'].str.contains(actor, na=False)]#filtra registros que contenga la cadena de texto ingresada
    result1 =result1[['title','release_year','type']]#sobreescribir reult1 con las columnas'title','release_year','type'
    result2 =result2[['title','release_year','type']]#sobreescribir reult2 con las columnas'title','release_year','type'
    result = pd.concat([result1,result2], axis=0).sort_values(by='release_year', ascending=True)#se
    result['type'] =result['type'].str.replace('Movie','Pelicula').replace('TV Show','Serie')
    return result.rename(columns={'title':'Título','release_year':'Año de Lanzamiento','type':'Tipo'}).reset_index(drop=True) 
```
